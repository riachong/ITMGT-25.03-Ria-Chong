```python
# write code here
months = {"1":"January","2":"February","3":"March","4":"April","5":"May","6":"June","7":"July","8":"August","9":"September","10":"October","11":"November","12":"December",
         "Jan":"January","Feb":"February","Mar":"March","Apr":"April","Jun":"June","Jul":"July","Aug":"August","Sep":"September","Oct":"October","Nov":"November","Dec":"December",
         "January":"January","February":"February","March":"March","April":"April","May":"May","June":"June","July":"July","August":"August","September":"September","October":"October","November":"November","December":"December"
         }

def clean_month(month):
    
    for key in months.keys():
        if month == key:
            return months[key]
        else:
            pass
```

```python
recweek = df.copy()
recweek["birthmonth"]=recweek["birthmonth"].apply(clean_month)
recweek
```

```python
# Clean Year

# dump recweek data
recweek
```

```python
# write code here
# Since 2020 is the max year, we attach "19" or "20" depending on the last two digits of the year

def clean_year(year):
    if int(str(year)[-2:]) > 20:
        year = str(year)
        if len(year) == 2:
            new_year = "19" + year
            return new_year
        else:
            return year
    elif int(str(year)[-2:]) <= 20:
        year = str(year)
        if len(year) == 2:
            new_year = "20" + year
            return new_year
        else:
            return year
```

```python
recweek["birthyear"]=recweek["birthyear"].apply(clean_year)
recweek
```

```python
# write code here
courses={"Management":"BS Management","MGT":"BS Management","BS MGT":"BS Management", 
        "Legal Management":"BS Legal Management","LM":"BS Legal Management","BS LM":"BS Legal Management", 
        "Management Engineering":"BS Management Engineering","ME":"BS Management Engineering","BS ME":"BS Management Engineering", 
        "Communications Technology":"BS Communications Technology","CTM":"BS Communications Technology","BS CTM":"BS Communications Technology",
        "Information Technology Entrepreneurship":"BS IT Entrepreneurship","ITE":"BS IT Entrepreneurship","BS ITE":"BS IT Entrepreneurship"
        }

def clean_course(course):
    
    for key in courses.keys():
        if course == key:
            return courses[key]
        else:
            pass
```

```python
recweek["course"]=recweek["course"].apply(clean_course)
recweek
```

```python
# mobile number cleaning

# dump recweek data
recweek
```

```python
def clean_number(number):
    new = re.findall(r"9[0-9]{2} [0-9]{3} [0-9]{4}",number)
    final = new[0]
    final = "0" + str(final)
    final = final.replace("+63 ","")
    return final
```

```python
recweek["phonenumber"] = recweek["phonenumber"].apply(clean_number)
recweek
```
