```python
#1 Turning Tables

n = int(input("Enter number of rows: "))
print("<table>")
for i in range(1,n+1):
    if i%2 == 0:
        print("<tr class=\"whiterow\"><td>",str(i),"</td><td>",str((i)**2),"</td></tr>", sep="")  
    else:
        print("<tr class=\"lightyellowrow\"><td>",str(i),"</td><td>",str((i)**2),"</td></tr>", sep="")
        
print("</table>")
```

```python
#4 Go Forth and Multiply

def multtable(x):
    for i in range(1,x+1,1):
        for j in range(1,x+1,1):
            print(i*j, end="\t")
        print() 
        
multtable(8)
```
